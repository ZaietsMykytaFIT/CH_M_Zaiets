Завдання: Побудувати інтерполяційний багаточлен Лагранжа (Варіант 3)

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import lagrange

x = np.array([-2, -1, 1, 2])  #Чомусь, без float все працює нормально
y = np.array([-26, -5, 1, 10])
x_4test = np.array([-0.5, 0.5, 1.5, 2.5])

def lagrange_interpolation(x, y, x_test):
    n = len(x)
    f_interp = np.zeros_like(x_test)

    for i_xt, x_test in enumerate(x_test):
        p = np.zeros(n)
        for i in range(n):
            p_i = 1
            for j in range(n):
                if i != j:
                    p_i *= (x_test - x[j]) / (x[i] - x[j])
            p[i] = p_i
        f_interp[i_xt] = np.dot(y, p)

    return f_interp

f_interp = lagrange_interpolation(x, y, x_4test)

print("Значення функції в точках x_4test:")
for i, x_i in enumerate(x_4test):
  print("x =",x_i, "\tf(x) = ", round(f_interp[i], 4))
```

```python
xnew = np.linspace(np.min(x), np.max(x), 100)
ynew = lagrange_interpolation(x, y, xnew)

fig = plt.figure(figsize = (6,3))
plt.plot(x, y, 'o', xnew, ynew)
plt.title('Lagrange Polynomial_1')
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
plt.show()

f = lagrange(x, y)    #Перевірка
fig = plt.figure(figsize = (6,3))
plt.plot(xnew, f(xnew), 'b', x, y, 'ro')
plt.title('Lagrange Polynomial_2')
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
